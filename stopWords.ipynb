## Removing stop words (the,on,in)

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
paragraph = '''Time management is one of the most essential skills a person can develop to achieve success in both personal and professional life. By organizing and prioritizing tasks effectively, individuals can make better use of their limited hours each day. Good time management not only reduces stress but also enhances productivity and focus. It allows people to set realistic goals, avoid procrastination, and maintain a healthy work-life balance. In today’s fast-paced world, where distractions are constant, mastering time management can make the difference between struggling to keep up and consistently reaching one’s full potential.'''

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(paragraph)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)

sentences
```
